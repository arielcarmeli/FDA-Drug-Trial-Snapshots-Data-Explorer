# Parse FDA diversity data - by yearly report

Annual Snapshots seen here - https://www.fda.gov/drugs/drug-approvals-and-databases/drug-trials-snapshots

```python
import pandas as pd
import numpy as np
```

```python
drug_phrases = []

with open('Raw_fda_trials_snapshot_2016.txt', 'r') as infile:
    for line in infile:
        split_line = line.split()
        #print(line)
        phrase_starts = []
        phrase_starts.append(0)
        
        # Find indexes for start of new drug
        for i in range(len(split_line)-1):        
            # If current is number and next is not a number, then next i is start of a new phrase
            if split_line[i].isnumeric() == True and split_line[i+1].isnumeric() == False: 
                if split_line[i+1] != "N/A": #"NR":
                    phrase_starts.append(i+1)
                    #print(i) 
        #print(phrase_starts)
                    
        # Create strings representing each drug (name, description, + 7 diversity values)
        for i in range(len(phrase_starts)):
            start_index = phrase_starts[i]
            #print("start:", start_index)
            drug_phrase = ""
            
            # If not last phrase
            if i < len(phrase_starts) - 1:
                for j in range(phrase_starts[i+1]-start_index):
                    #print(j)
                    #print(split_line())
                    drug_phrase += str(split_line[phrase_starts[i]+j]) + " "
            
            # If last phrase
            else:
                #print("last", start_index)
                #print("size", len(split_line)-1-start_index)
                for j in range(len(split_line)-start_index):
                    #print(j)
                    drug_phrase += str(split_line[phrase_starts[i]+j]) + " "                   
            
            drug_phrases.append(drug_phrase) 
            #print(drug_phrase)
            
drug_phrases
```
